```python
import numpy as np
import pandas as pd
```

```python
## load the PPC Campaign Performance Dataset
df = pd.read_excel(r"C:\Users\aluis\OneDrive\Desktop\PPC_Data\ppc_campaign_performance_data.xlsx")
```

```python
### peek into the data
df.head(20)
```

```python
# get overview of dataset structure
df.info()
```

```python
## check for missing values
df.isnull().sum()
```

```python
# now we standardize our column names
# we will remove leading spaces, convert to lowercase, and replace spaces with underscores for column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df
```

```python
# remove leading spaces from data values in dataset
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
df
```

```python
# Convert date column from object data type to datetime data type and validate dates
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# check to see number of invalid dates after conversion
df['date'].isna().sum()
```

```python
# converting date back to string for MYSQL data import
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
```

```python
df
```

```python
# check for duplicate rows
df.duplicated().sum()
```

```python
# resetting index
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# Export clean file as CSV for MYSQL
df.to_csv("ppc_campaign_cleaned.csv", index=False)
```
